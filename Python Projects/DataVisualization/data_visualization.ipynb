### Problem 1

The file `anscombe.npy` contains Anscombe’s quartet.
For each section of the quartet,
- Plot the data as a scatter plot on the box [0, 20] × [0, 13].
- Use `scipy.stats.linregress()` to calculate the slope and intercept of the least squares regression line for the data and its correlation coefficient (the first three return values).
- Plot the least squares regression line over the scatter plot on the domain $x\in[0,20]$.
- Report the mean and variance in $x$ and $y$, the slope and intercept of the regression line, and the correlation coefficient. Compare these statistics to those of the other sections.
- Describe how the section is similar to the others and how it is different.

```python
A = np.load("anscombe.npy")
A1 = A[:,0:2]
A2 = A[:,2:4]
A3 = A[:,4:6]
A4 = A[:,6:8]


slope, intercept, r_value, p_value, std_err = stats.linregress(A1[:,0], A1[:,1])
x = np.linspace(0, 20, 50)

plt.scatter(A1[:,0], A1[:,1])
plt.plot(x, slope * x + intercept)
plt.axis([0, 20, 0, 13])
plt.show()

xmean, ymean = np.mean(A1[:,0]), np.mean(A1[:,1])
xvar, yvar = np.var(A1[:,0]), np.mean(A1[:,1])
print("Means: ", xmean, ymean, "\nVariance: ", xvar, yvar, "\nSlope, intercept, coefficient: ", slope, intercept, r_value)

slope, intercept, r_value, p_value, std_err = stats.linregress(A2[:,0], A2[:,1])

plt.scatter(A2[:,0], A2[:,1])
plt.plot(x, slope * x + intercept)
plt.axis([0, 20, 0, 13])
plt.show()

xmean, ymean = np.mean(A2[:,0]), np.mean(A2[:,1])
xvar, yvar = np.var(A2[:,0]), np.mean(A2[:,1])
print("Means: ", xmean, ymean, "\nVariance: ", xvar, yvar, "\nSlope, intercept, coefficient: ", slope, intercept, r_value)

slope, intercept, r_value, p_value, std_err = stats.linregress(A3[:,0], A3[:,1])

plt.scatter(A3[:,0], A3[:,1])
plt.plot(x, slope * x + intercept)
plt.axis([0, 20, 0, 13])
plt.show()

xmean, ymean = np.mean(A3[:,0]), np.mean(A3[:,1])
xvar, yvar = np.var(A3[:,0]), np.mean(A3[:,1])
print("Means: ", xmean, ymean, "\nVariance: ", xvar, yvar, "\nSlope, intercept, coefficient: ", slope, intercept, r_value)

slope, intercept, r_value, p_value, std_err = stats.linregress(A4[:,0], A4[:,1])

plt.scatter(A4[:,0], A4[:,1])
plt.plot(x, slope * x + intercept)
plt.axis([0, 20, 0, 13])
plt.show()

xmean, ymean = np.mean(A4[:,0]), np.mean(A4[:,1])
xvar, yvar = np.var(A4[:,0]), np.mean(A4[:,1])
print("Means: ", xmean, ymean, "\nVariance: ", xvar, yvar, "\nSlope, intercept, coefficient: ", slope, intercept, r_value)

print("Although each subplot of data is different, the least square best fit line and variance are identical for all 4 sets of data.")

```

### Problem 2
The $n + 1$ Bernstein basis polynomials of degree $n$ are defined as follows.

$$b_{n,v}(x) = \binom{n}{v}x^v(1-x)^{n-v},\quad v=0,\ 1,\ \ldots,\ n$$

Plot the first $10$ Bernstein basis polynomials ($n = 0,\ 1,\ 2,\ 3$) as small multiples on the domain $[0, 1] \times [0, 1]$.
Label the subplots for clarity, adjust tick marks and labels for simplicity, and set the window limits of
each plot to be the same.
Consider arranging the subplots so that the rows correspond with $n$ and the columns with $v$.

(Hint: The constant $\binom{n}{v} = \frac{n!}{v!(n-v)!}$ is called the binomial coefficient and can be efficiently computed with `scipy.special.binom()` or `scipy.misc.comb()`.)

```python
plt.axis([0, 1, 0, 1])
x = np.linspace(0, 1, 200)
r = 4

ax1 = plt.subplot(221)
n0 = 0
v = 0
ax1.plot(x, special.binom(n0, v) * x ** v * (1 - x) ** (n0-v), label=r"$B_{}(x)$".format(v))
    
ax1.legend(loc="right")
ax1.set_title("n=0")
    
ax2 = plt.subplot(222)
n1 = 1
for v in range(n1 + 1):
    ax2.plot(x, special.binom(n1, v) * x ** v * (1 - x) ** (n1-v), label=r"$B_{}(x)$".format(v))
    
ax2.legend(loc="right")
ax2.set_title("n=1")
    
ax3 = plt.subplot(223)
n2 = 2
for v in range(n2 + 1):
    ax3.plot(x, special.binom(n2, v) * x ** v * (1 - x) ** (n2-v), label=r"$B_{}(x)$".format(v))
    
ax3.legend(loc="right")
ax3.set_title("n=2")

ax4 = plt.subplot(224)
n3 = 3
for v in range(n3 + 1):
    ax4.plot(x, special.binom(n3, v) * x ** v * (1 - x) ** (n3-v), label=r"$B_{}(x)$".format(v))
    
ax4.legend(loc="right")
ax4.set_title("n=3")
        
plt.show()



```

### Problem 3

The file `MLB.npy` contains measurements from over 1,000 recent Major League Baseball players, compiled by UCLA.
Each row in the array represents a different player; the columns are the player’s height (in inches), weight (in pounds), and age (in years), in that order.

Create several visualizations to show the correlations between height, weight, and age in the MLB data set.
Use at least one scatter plot.
Adjust the marker size, plot a regression line, change the window limits, and use small multiples where appropriate.

```python
height, weight, age = np.load("MLB.npy").T

ax1 = plt.subplot(131)
ax1.set_xlabel("Height")
ax1.hist(height, bins=50)

ax2 = plt.subplot(132)
ax2.set_xlabel("Weight")
ax2.hist(weight, bins=50)

ax3 = plt.subplot(133)
ax3.set_xlabel("Age")
ax3.hist(age, bins=50)
plt.show()

x = np.linspace(65, 85, 100)
plt.scatter(height, weight, alpha=.5, edgecolor='none')
slope, intercept, r_value, p_value, std_err = stats.linregress(height, weight)
plt.plot(x, slope * x + intercept)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()

x = np.linspace(18, 50, 300)
plt.scatter(age, height, alpha=.5, edgecolor='none')
slope, intercept, r_value, p_value, std_err = stats.linregress(age, height)
plt.plot(x, slope * x + intercept)
plt.xlabel("Age")
plt.ylabel("Height")
plt.show()



```

### Problem 5

The _Rosenbrock function_ is defined as follows. $$f(x,y)=(1−x)^2 +100(y−x^2)^2$$
The minimum value of $f$ is $0$, which occurs at the point $(1, 1)$ at the bottom of a steep, banana-shaped valley of the function.

Use a heat map and a contour plot to visualize the Rosenbrock function.
Plot the minimizer $(1, 1)$.
Use a different sequential colormap for each visualization.

```python
x = np.linspace(0, 2, 200)
X, Y = np.meshgrid(x, x)
Z = (1 - X) ** 2 + 100 * (Y - X ** 2) ** 2
plt.contourf(X, Y, Z, 10, cmap="viridis")
plt.plot(1, 1, 'co')
plt.show()

plt.pcolormesh(X, Y, Z, cmap="inferno", shading='auto')
plt.plot(1, 1, 'mo')
plt.show()
```

### Problem 6

The file `countries.npy` contains information from 20 different countries.
Each row in the array represents a different country; the columns are the 2015 population (in millions of people), the 2015 GDP (in billions of US dollars), the average male height (in centimeters), and the average female height (in centimeters), in that order.

Visualize this data set with at least four plots, using at least one scatter plot, one histogram, and one bar chart.
List the major insights that your visualizations reveal.

(Hint: consider using `np.argsort()` and fancy indexing to sort the data for the bar chart.)

```python
countries = np.array(
            ["Austria", "Bolivia", "Brazil", "China", "Finland",
             "Germany", "Hungary", "India", "Japan", "North Korea",
             "Montenegro", "Norway", "Peru", "South Korea", "Sri Lanka",
             "Switzerland", "Turkey", "United Kingdom", "United States", "Vietnam"])
population, gdp, male, female = np.load("countries.npy").T

plt.scatter(male, female)
plt.title("Average male and female height per country")
plt.xlabel("Male Height")
plt.ylabel("Female Height")
x = np.linspace(160, 185, 50)
slope, intercept, r_value, p_value, std_err = stats.linregress(male, female)
plt.plot(x, slope * x + intercept)

plt.show()

plt.hist(gdp, bins=45)
plt.title("Gross Domestic Product")
plt.xlabel("Country GDP (in billions USD)")
plt.ylabel("Number of countries")
plt.show()

positions = np.arange(len(countries))
plt.barh(positions, population, align="center")
plt.yticks(positions, countries)
plt.xlabel("Population (millions)")
plt.show()

plt.scatter(male, population)
plt.ylabel("Population (millions)")
plt.xlabel("Average Male Height (cm)")
plt.title("Average male height compared to country population")
plt.show()
```

From Graph 1: Female Height and Male Height tend to trend in the same direction
From Graph 2: Most countries have a GDP somewhere in the range of 0 to 500 billion, with 2 countries being outliers at above 10000 billion
From Graph 3: China and India are by far the most populous countries, with the US coming in a distanct third.
From Graph 4: There seems to be very little, if any, correlation between male height and total population.
